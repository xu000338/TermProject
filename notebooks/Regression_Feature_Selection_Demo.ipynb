# Linear Regression with Multiple Features

### Using 2 features predict `salary`

```python
# import pandas
import pandas as pd

#import the dataset
df = pd.read_csv('Pay.csv')
df.shape
```

```python
# find correlations
df.corr()
```

### Feature Selection


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.scatter(df.Age, df.Salary)
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('Age vs Salary')


plt.subplot(2,2,2)
plt.scatter(df.Experience, df.Salary)
plt.xlabel('Experience')
plt.ylabel('salary')
plt.title('Experience vs Salary')
plt.show()
```

## Now let's build a simple linear regression model

We will use the variable `Age` to predict `Salary`

```python
df.head()
```

```python
# Seperate features and target
x = df[['Age']]
y = df[['Salary']]
```

```python
# import module
from sklearn.model_selection import train_test_split

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=123)
```

```python
plt.scatter(x_train,y_train,color='lightgray')
plt.scatter(x_test, y_test, color='red')
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('Age vs Salary')
plt.legend(['Train data points','Test data points'], loc='lower right')
plt.show()
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# import the model
from sklearn.linear_model import LinearRegression

# Create an instance of the model
Regression = LinearRegression()
```

```python
# train the model
model = Regression.fit(x_train,y_train)
```

```python
# Make prediction on train set
y_trainpred = model.predict(x_train)

# make prediction on test set
y_pred = model.predict(x_test)
```

```python
# get the parameters
print('Model coefficient is', model.coef_)
print('Model intercept is', model.intercept_)
```

```python
ddf = pd.DataFrame()
ddf['x_test'] = x_test
ddf['y_pred'] = y_pred
ddf['y_test'] = y_test
```

```python
ddf
```

```python
from sklearn.metrics import mean_squared_error

# evaluate the model
rmse1 = mean_squared_error(y_trainpred, y_train, squared=False)
rmse2 = mean_squared_error(y_pred, y_test, squared=False)
print('rmse on train set is',rmse1)
print('rmse on test set is', rmse2)
```
